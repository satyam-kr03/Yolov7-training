```python
# open ./annotations.csv

import pandas as pd
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# open and convert to dataframe
df = pd.read_csv('./annotations.csv')
df = pd.DataFrame(df)

print(df['image'])
```

```python
# for every value in the image column, search for the image in the images folder (./training_images)
# print a message if the image is not found

# for index, row in df.iterrows():
#     image_name = row['image']
#     image_path = os.path.join('./training_images', image_name)
#     if not os.path.exists(image_path):
#         print(f"Image {image_name} not found in training_images folder.")
#     else:
#         print(f"Image {image_name} found in training_images folder.")

# now perform a more refined search by ignoring the last 4 characters of the image name (extension may be different)

for index, row in df.iterrows():
    image_name_no_ext = row['image'][:-4]  # Remove extension
    found = False
    for file in os.listdir('./training_images'):
        if file.startswith(image_name_no_ext):
            found = True
            # replace the image name in the dataframe with the file name
            df.at[index, 'image'] = file
            break
    if not found:
        print(f"Image {row['image']} not found in training_images folder (by name without extension).")
        # drop the row from the dataframe
        df.drop(index, inplace=True)
        print(f"Dropping row with index {index}.")

# save the dataframe to a new csv file
df.to_csv('./annotations_updated.csv', index=False)
```
