```python
ANNOTATIONS_DIR = "./CSV_GT"
TARGET_DIR = "./new_data/"

import os
import pandas as pd
import numpy as np

# initialize target empty dataframe
target_df = pd.DataFrame(columns=["image", "xmin", "ymin", "xmax", "ymax", "class_name"])

# iterate over all .csv files in the directory
for filename in os.listdir(ANNOTATIONS_DIR):
    if filename.endswith(".csv"):
        # read the csv file
        df = pd.read_csv(os.path.join(ANNOTATIONS_DIR, filename))

        # iterate over each row in the dataframe
        for index, row in df.iterrows():
            # extract the relevant information
            image_name = filename.replace('.csv', '.jpg')
            l = row['l']
            t = row['t']
            w = row['w']
            h = row['h']
            class_name = row['class_name'].lower()

            '''
            The format is:
                l - pixel number from left of image
                t - pixel number from top of image
                w - width of bounding box
                h - height of bounding box

            We need to convert these to:
                xmin - pixel number from left of image
                ymin - pixel number from top of image
                xmax - pixel number from right of image
                ymax - pixel number from bottom of image
            '''

            # calculate the bounding box coordinates
            xmin = l
            ymin = t
            xmax = l + w
            ymax = t + h

            # create a new dataframe with the extracted information
            new_row = pd.DataFrame({
                "image": [image_name],
                "xmin": [xmin],
                "ymin": [ymin],
                "xmax": [xmax],
                "ymax": [ymax],
                "class_name": [class_name]
            })

            # append the new dataframe to the target dataframe
            target_df = pd.concat([target_df, new_row], ignore_index=True)

            # break

# save the target dataframe to a new CSV file
target_df.to_csv(os.path.join(TARGET_DIR, "annotations.csv"), index=False)
# print the target dataframe
print(target_df)
    
```
